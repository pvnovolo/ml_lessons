# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset('tips')
```

### Посмотрите на первые 5 строчек

```python
tips.head(5)
```

### Узнайте сколько всего строчек и колонок в данных

```python
print("Количество строк и столбцов:", tips.shape)
```

### Проверьте есть ли пропуски в данных

```python
print(tips.isnull().sum())
```

### Посмотрите на распределение числовых признаков

```python
import matplotlib.pyplot as plt
# Построение гистограммы для признака "total_bill"
sns.histplot(data=tips, x="total_bill")
plt.show()

# Построение гистограммы для признака "tip"
sns.histplot(data=tips, x="tip")
plt.show()

# Построение гистограммы для признака "size"
sns.histplot(data=tips, x="size")
plt.show()
```

### Найдите максимальное значение 'total_bill'

```python
max_total_bill = tips['total_bill'].max()
print(max_total_bill)
```

### Найдите количество курящих людей

```python
smokers_count = tips['smoker'].value_counts()['Yes']
print(smokers_count)
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
mean_total_bill_by_day = tips.groupby('day')['total_bill'].mean()
print(mean_total_bill_by_day)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_total_bill = tips['total_bill'].median()
tips_above_median = tips.loc[tips['total_bill'] > median_total_bill]
mean_tip_by_sex = tips_above_median.groupby('sex')['tip'].mean()
print(mean_tip_by_sex)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = tips['smoker'].replace({'No': 0, 'Yes': 1})
print(tips.head())
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
tips['total_bill'].hist()
plt.xlabel('Total bill')
plt.ylabel('Frequency')
plt.title('Distribution of total bill')
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
# scatterplot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title('Scatterplot of total bill and tip')
plt.show()

```

### Постройте pairplot

```python
sns.pairplot(tips)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
sns.scatterplot(x='total_bill', y='day', data=tips)
plt.xlabel('Total bill')
plt.ylabel('Day')
plt.title('Scatterplot of total bill and day')
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
sns.histplot(data=tips, x='tip', hue='time', multiple='stack')
plt.title('Histogram of tip by time')
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
sns.scatterplot(x='total_bill', y='tip', data=tips[tips['sex']=='Male'], hue='smoker')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title('Scatterplot of total bill and tip for Male')
plt.show()

sns.scatterplot(x='total_bill', y='tip', data=tips[tips['sex']=='Female'], hue='smoker')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title('Scatterplot of total bill and tip for Female')
plt.show()
```
