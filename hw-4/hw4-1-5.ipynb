```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy.linalg import svd, norm
```

```python
n=100
```

```python
A = np.random.random((n,n))
```

```python
# Сингулярное разложение матрицы
U,S,Vt = svd(A)
# print(U)
# print(S)
# print(Vt)
```

```python
rank = list(range(2,n+1))
errors =[]
for r in rank:
    U_r = U[:,:r]
    S_r = S[:r]
    Vt_r = Vt[:r,:]
    A_r = U_r @ np.diag(S_r) @ Vt_r
    er = norm(A-A_r)
    errors.append(er)
print(errors)
```

```python
# Строим график зависимости ошибки от ранга
plt.figure(figsize=(10,6))
plt.plot(rank, errors,label='Ошибка аппроксимации')
plt.xlabel('Ранг матрицы')
plt.ylabel('Ошибка')
plt.title('Зависимость ошибки аппроксимации от ранга матрицы')
plt.legend()
plt.grid(True)
plt.show()
```
